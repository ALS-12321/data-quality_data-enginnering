# Examenes realizados de laboratorio

## 1.-Cargar y visualizar el archivo

```python
import pandas as pd
df = pd.read_csv('Laboratory_Examinations_Conducted.txt', delimiter=',')
print(df.head())
print(df.dtypes)
```

## 2.-Mostrar tipos de datos "Algunas columnas, como la fecha del examen y el costo, podrían no ser reconocidas automáticamente como numéricas debido a errores en los datos (como valores negativos o faltantes)".

```python
print("Precios negativos:")
print(df[df["Precio (USD)"] <= 0])

print("Fecha de examen inválidos:")
print(df[df["Fecha del Examen"].isna()])
```

## 3.-Calcular estadísticas (media, mediana, desviación estándar) "Una vez que las columnas estén en formato numérico, calcularemos las estadísticas solicitadas"

```python
# Calcular media, mediana y desviación estándar de "Precio" y "Resultado"
precio_media = df['Precio (USD)'].mean()
precio_mediana = df['Precio (USD)'].median()
precio_desviacion_std = df['Precio (USD)'].std()

resultado_media = df['Resultado'].mean()
resultado_mediana = df['Resultado'].median()
resultado_desviacion_std = df['Resultado'].std()

print("Estadísticas del Precio:")
print(f"Media: {precio_media}")
print(f"Mediana: {precio_mediana}")
print(f"Desviación Estándar: {precio_desviacion_std}")

print("\nEstadísticas de Resultados:")
print(f"Media: {resultado_media}")
print(f"Mediana: {resultado_mediana}")
print(f"Desviación Estándar: {resultado_desviacion_std}")
```

## 4.-Crear visualizaciones de los datos "Ahora generaremos histogramas de las columnas numéricas para visualizar las distribuciones del precio y resultado de examen"

```python
import matplotlib.pyplot as plt
# Graficar histogramas de preio y Resultado 
plt.figure(figsize=(14, 6))

# Histograma de Precio
plt.subplot(1, 2, 1)
plt.hist(df['Precio (USD)'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de Precios')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')

# Histograma de Resultado
plt.subplot(1, 2, 2)
plt.hist(df['Resultado'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Histograma de Resultado')
plt.xlabel('Resultados de examenes')
plt.ylabel('Frecuencia')

# Mostrar las gráficas
plt.tight_layout()
plt.show()
```
